**Q1. What is hypothesis testing in statistics?**

**Ans:** Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on sample data. It helps us check whether there is enough evidence in the sample to support or reject a claim (hypothesis) about the population.

Here’s the breakdown:

**1. Formulate Hypotheses:**

Null hypothesis (H₀): A default assumption (e.g., there is no effect, no difference).

Alternative hypothesis (H₁ or Ha): What we want to test (e.g., there is an effect, a difference).


For example:

**H₀**: The average height of students = 165 cm.

**H₁**: The average height of students ≠ 165 cm.




**2. Choose Significance Level (α):**

This is the probability of rejecting H₀ when it is actually true (a Type I error).
Common choices: 0.05 (5%), 0.01, or 0.10.



**3. Select a Test & Calculate Test Statistic:**

Depending on the data and situation, use a statistical test (z-test, t-test, chi-square test, ANOVA, etc.).
The test statistic measures how far the sample result is from what H₀ predicts.



**4. Find the p-value:**

p-value = probability of getting the observed (or more extreme) result if H₀ is true.

If p ≤ α, reject H₀ (evidence supports H₁).

If p > α, fail to reject H₀ (not enough evidence against H₀).




**5. Make a Conclusion:**

Reject H₀: There is enough evidence to support the alternative hypothesis.

Fail to reject H₀: Not enough evidence to support the alternative; we stick with H₀.




**In short:**

* Hypothesis testing is like a courtroom trial.

* H₀ = defendant is innocent.

* Evidence (data) is presented.

* If evidence is strong enough (p ≤ α), reject H₀ → guilty (support H₁).

* If not, fail to reject H₀ → not guilty.

**Q2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?**

Ans:

 **Null Hypothesis (H₀):**

* The null hypothesis is the default or status quo assumption.

* It states that there is no effect, no difference, or no relationship in the population.

* Example: The new medicine has no effect on blood pressure.




**Alternative Hypothesis (H₁ or Ha):**

* The alternative hypothesis is what we want to test or prove.

* It states that there is an effect, a difference, or a relationship in the population.

* Example: The new medicine lowers blood pressure.

**Q3. Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.**

**Ans: The significance level (α):**

The significance level (denoted by α) is the threshold we set before performing a hypothesis test.
It represents the probability of making a Type I error → rejecting the null hypothesis (H₀) when it is actually true.


Common choices: 0.05 (5%), 0.01 (1%), 0.10 (10%).



**ROLE IN HYPOTHESIS TESTING**

**1. Compare p-value with α:**

If p ≤ α → reject H₀ (evidence supports H₁).

If p > α → fail to reject H₀ (not enough evidence against H₀).



**2. Controls Risk of Error:**

Lower α (e.g., 0.01) → very strict, less chance of false positives, but more chance of missing real effects (Type II error).

Higher α (e.g., 0.10) → more lenient, more chance of detecting effects, but also higher risk of false positives.





**For example:**

Suppose we test whether a coin is fair:

H₀: The coin is fair (p = 0.5).

H₁: The coin is biased (p ≠ 0.5).

Choose α = 0.05.


* If the test gives p = 0.03 → since 0.03 < 0.05, we reject H₀ → conclude the coin is biased.

* If the test gives p = 0.12 → since 0.12 > 0.05, we fail to reject H₀ → not enough evidence to say the coin is biased.


##Mainly:
The significance level (α) is like a cutoff line. It defines how much evidence is required to reject H₀ and helps balance between detecting true effects and avoiding false alarms.



**Q5. What is the difference between a Z-test and a T-test? Explain when to use each.**

**Ans:
Z-test:**

A statistical test used when:

1. The population variance (σ²) is known.


2. The sample size is large (n ≥ 30).



Based on the standard normal distribution (Z-distribution).

Commonly used for proportions or when dealing with very large samples.


For example:
Testing whether the average height of 10,000 students is 165 cm when the population variance is already known.




**T-test:**

A statistical test used when:

1. The population variance is unknown.


2. The sample size is small (n < 30).



Based on the t-distribution, which is wider and accounts for extra uncertainty in small samples.

As n increases, the t-distribution approaches the normal distribution (so large-sample t-test ≈ z-test).


For example:
Testing whether the average exam score of a class of 20 students is 70, without knowing the population variance.

**In other words:**

* Use Z-test → when population variance is known & sample size is large.

* Use T-test → when population variance is unknown & sample size is small.

**Q5. What is the difference between a Z-test and a T-test? Explain when to use each.**

**Ans:** The key difference between a Z-test and a T-test lies in their assumptions about the population **standard deviation**.

**1. Z-test:**
*  Population standard deviation (σ) is **known**.
* Generally used when the sample size is **large** (n ≥ 30)
* Uses the **standard normal (Z) distribution**.
* Critical values are **fixed**(eg. ±1.96 for 95% confidence).
* Often used in quality control, large-sample survey or when past data gives a reliable σ.
**2. T-test:**
* Population standard deviation is **unknown**.
* Suitable for **small samples** (n<30) but can be used for any size if σ is unknown.
* Uses **Student's t-distribution**, which has heavier tails.
* Critical values **depends on the sample size** via degrees of freedom.
* Common in research studies where σ is not known, such as medical or social science experiments.

**WHEN TO USE:**

*Use Z-test when:*
* Population standard deviation σ is known
* Sample size is large (n ≥ 30) so the Central Limit Theorem ensures the sample mean is approximately normal.

Typical situations: quality-control checks, large surveys, or industrial processes where historical data provide a reliable σ.

*Use T-test when: *
* The population standard deviation is unknown (which is common), and

* The sample size is small (n < 30) or of any size if σ is not known, provided the data are roughly normal.

Typical situations: medical or social-science experiments, comparing means of two small groups, or before/after (paired) studies.

**Q6. Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10  # Number of trials
p = 0.5 # Probability of success

# Generate random numbers from a binomial distribution
# The 'size' parameter determines how many random numbers to generate
num_samples = 1000
random_numbers = np.random.binomial(n, p, size=num_samples)

# Create a histogram
plt.hist(random_numbers, bins=np.arange(n + 2) - 0.5, density=True, rwidth=0.8, color='skyblue', edgecolor='aqua')
plt.title(f'Binomial Distribution Histogram (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(range(n+1))
plt.grid(axis='y', alpha=0.75)
plt.show()
```

**Q7. Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]**


```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypotheses
mu_0 = 50.0            # hypothesized population mean
sigma = 0.5            # known population std dev
alpha = 0.05           # significance level

# Sample statistics
x_bar = np.mean(sample_data)
n = len(sample_data)

# Z statistic
z_stat = (x_bar - mu_0) / (sigma / np.sqrt(n))

# p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample mean: {x_bar:.3f}")
print(f"Sample size: {n}")
print(f"Z statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Decision
if p_value < alpha:
    print("Reject H0: Significant evidence that the mean differs from 50.")
else:
    print("Fail to reject H0: No significant evidence that the mean differs from 50.")
```

**Q8. Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm

# Simulate data from a normal distribution
np.random.seed(42)                       # for reproducibility
mu_true = 100                              # true mean
sigma_true = 15                             # true standard deviation
n = 50                                    # sample size

data = np.random.normal(mu_true, sigma_true, n)
# Calculate the sample mean and standard error
sample_mean = np.mean(data)
standard_error = stats.sem(data)

#  Calculate 95% confidence interval for the mean
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)          # sample standard deviation
confidence_level = 0.95
alpha = 0.05
z_critical = norm.ppf(1 - alpha/2)         # Z critical value for 95%

margin_error = z_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_error
ci_upper = sample_mean + margin_error

print(f"Sample mean: {sample_mean:.2f}")
print(f"95% CI for mean: [{ci_lower:.2f}, {ci_upper:.2f}]")

#  Plot the confidence interval
plt.figure(figsize=(8,5))
plt.hist(data, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label='Sample Mean')
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label='95% CI Lower')
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label='95% CI Upper')
plt.title("Histogram of Simulated Normal Data with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.grid(axis='y', alpha=0.75)
plt.show()
```

**Q9. Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.**

```python
import numpy as np
import matplotlib.pyplot as plt

def zscore_and_plot(data, bins=20):
    """
    Calculate Z-scores for a 1-D dataset and plot their histogram.

    Parameters
    ----------
    data : array-like
        Input numerical dataset.
    bins : int
        Number of bins for the histogram.

    Returns
    -------
    z_scores : np.ndarray
        Array of Z-scores for each observation.
    """
    data = np.asarray(data)
    mean = np.mean(data)
    std = np.std(data, ddof=0)   # population std; use ddof=1 for sample
    z_scores = (data - mean) / std

    print(f"Mean of original data : {mean:.2f}")
    print(f"Standard deviation    : {std:.2f}")
    print(f"First 5 Z-scores      : {z_scores[:5]}")

    # Plot histogram of Z-scores
    plt.figure(figsize=(7,4))
    plt.hist(z_scores, bins=bins, color="skyblue", edgecolor="black", alpha=0.7)
    plt.axvline(0, color="red", linestyle="dashed", linewidth=2, label="Mean (0)")
    plt.title("Histogram of Z-scores (Standardized Data)")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.legend()
    plt.tight_layout()
    plt.show()

    return z_scores

# Example usage with simulated data
if __name__ == "__main__":
    np.random.seed(0)
    sample = np.random.normal(loc=100, scale=15, size=200)
    z_scores = zscore_and_plot(sample)
```
